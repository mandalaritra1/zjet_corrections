```python
import pickle as pkl
```

```python
filename = 'outputs/st_all.pkl'
with open(filename, 'rb') as f:
    out = pkl.load(f)
```

```python
out.keys()
```

```python
new_dict = {}
for hname, h in out.items():
    if hname not in ['sumw', 'nev', 'cutflow']:
        new_dict[hname] = h[..., "nominal"]
    else:
        new_dict[hname] = h
        
    
```

```python
outfile = 'outputs/st_all_trimmed.pkl'
with open(outfile, 'wb') as f:
    pkl.dump(new_dict, f)
```
